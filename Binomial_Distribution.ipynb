```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10 # number of trials
p = 0.5 # probability of success
size = 1000 # number of sample to generate

binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```
