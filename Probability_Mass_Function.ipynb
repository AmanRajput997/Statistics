**Single Dice Rolling...**

```python
import pandas as pd
```

```python
import random
```

```python
L = []
for i in range(10000):
  L.append(random.randint(1,6))
```

```python
len(L)
```

```python
L[:5]
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind = 'bar')
```

**Double Dice Rolling...**

```python
import pandas as pd
```

```python
import random
```

```python
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)

  L.append(a+b)
```

```python
len(L)
```

```python
L[:5]
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s
```

```python
s.plot(kind = 'bar')
```
