```python
import seaborn as sna
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame()
```

```python
x = pd.Series([12,25,68, 42,113])
y = pd.Series([11,29,58,121,100])
```

```python
df['x'] = x
df['y'] = y
```

```python
df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot scatterplot on each axes
ax1.scatter(df['x'], df['y'])
ax2.scatter(df['x']*2, df['y']*2)

ax1.set_title("Covariance - " + str(np.cov(df['x'],df['y'])[0,1]))
ax2.set_title("Correlation - " + str(np.corrcoef(df['x']*2,df['y']*2)[0,1]))

plt.tight_layout()
plt.show()
```
